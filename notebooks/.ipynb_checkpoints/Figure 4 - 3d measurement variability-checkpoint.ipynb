```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
working_dir = '/data/datasets/organoid_phenotyping/analysis/d35_vs_d60/'
os.listdir(working_dir)
```

```python
df = pd.read_excel(os.path.join(working_dir, 'combined_features.xlsx'), index_col=0)
list(df.keys())
```

```python
df = df.fillna(method='backfill')
df2 = df.drop(columns=['20190419_15_50_16_AA_org2_488LP13_561LP120_642LP60',
                      '20190614_18_12_44_AA_ef23-6_488LP14_561LP120_642LP50',
                      '20190509_16_55_31_AA-orgs5.8.19_org1_488LP15_561LP140_642LP50',
                      '20190531_14_31_36_AA_fatty2_488LP13_561LP140_642LP60',
                      '20190531_15_00_56_AA_fatty3_488LP13_561LP120_642LP50',
                      '20190531_15_57_33_AA_fatty1_488LP13_561LP100_642LP60',
                      '20190531_17_03_00_AA_fatty4_488LP13_561LP100_642LP50',
                      '20190531_17_54_24_AA_fatty5_488LP14_561LP100_642LP50',
                      '20190531_18_54_48_AA_fatty6_488LP14_561LP100_642LP50',
                      '20190531_20_32_55_AA_fatty7_488LP13_561LP140_642LP50'])

sns.clustermap(df2, z_score=0)
plt.show()
```

```python
for i, f in enumerate(df2.index):
    print(i, f)
```

```python
means = np.asarray([row.mean() for _, row in df2.iterrows()])
stdevs = np.asarray([row.std() for _, row in df2.iterrows()])
scale = 157 * ['singlecell'] + (217 - 157) * ['cyto'] + (275 - 217) * ['wholeorg']

df3 = pd.DataFrame({'feature': df2.index, 
                    'mean': means, 
                    'stdev': stdevs, 
                    'n': len(df2) * [len(list(df2.keys()))],
                    'scale': scale})
df3.head()
```

```python
df3['stderr'] = df3['stdev'] / np.sqrt(df3['n'])
df3.head()
```

```python
df3['rel_stderr'] = df3['stderr'] / df3['mean']
df3.head()
```

```python
df3_sort = df3.sort_values('rel_stderr', ascending=False)

display(df3_sort.head())
df3_sort.tail()
```

```python
df3_sort['feature']
```

```python
# Drop any DP of Artifact related features
keep = np.asarray(['DP' not in f and 'Artifact' not in f for f in df3_sort['feature']])
df3_sort_filt = df3_sort.iloc[keep]

len(df3_sort), len(df3_sort_filt)
```

```python
df3_sort_filt.head(20)
```

```python
df3_sort_filt.tail(20)
```

```python
sns.scatterplot(x=np.arange(len(df3_sort_filt)), 
                y='rel_stderr', 
                hue='scale',
                data=df3_sort_filt,
                edgecolor=None,
                s=8)
plt.xlabel('Feature')
plt.ylabel('Relative SE')
plt.ylim([0, 0.4])
sns.despine()
plt.show()
```

```python
sns.distplot(df3_sort_filt['rel_stderr'], kde_kws={'bw': 0.015})
sns.despine()
plt.show()
```
