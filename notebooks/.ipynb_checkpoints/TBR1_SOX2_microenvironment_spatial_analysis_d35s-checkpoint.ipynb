```python
start = np.asarray([0.5, 0.3])
stop = np.asarray([0.6, 0.4])

plt.figure(figsize=(9, 4))
ax1 = plt.subplot(1, 2, 1)
plt.scatter(prox_tbr1_mock[:, 0], prox_tbr1_mock[:, 1], s=1)
rect = Rectangle((start + stop)/2, stop[0]-start[0], stop[1]-start[1], edgecolor='r', linewidth=1, fill=False)
ax1.add_patch(rect)
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.title(f'TBR1+ cells in Mock_{mock_idx}')
plt.xlabel('SOX2 proximity')
plt.ylabel('TBR1 proximity')
ax2 = plt.subplot(1, 2, 2)
plt.scatter(prox_tbr1_zika[:, 0], prox_tbr1_zika[:, 1], s=1)
rect = Rectangle((start + stop)/2, stop[0]-start[0], stop[1]-start[1], edgecolor='r', linewidth=1, fill=False)
ax2.add_patch(rect)
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.title(f'TBR1+ cells in Zika_{zika_idx}')
plt.xlabel('SOX2 proximity')
plt.ylabel('TBR1 proximity')
plt.tight_layout()
plt.show()
```

```python
def show_gate(start, stop):
    plt.figure(figsize=(9, 4))
    ax1 = plt.subplot(1, 2, 1)
    plt.scatter(prox_tbr1_mock[:, 0], prox_tbr1_mock[:, 1], s=1)
    rect = Rectangle((start + stop)/2, stop[0]-start[0], stop[1]-start[1], edgecolor='r', linewidth=1, fill=False)
    ax1.add_patch(rect)
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.title(f'TBR1+ cells in Mock_{mock_idx}')
    plt.xlabel('SOX2 proximity')
    plt.ylabel('TBR1 proximity')
    ax2 = plt.subplot(1, 2, 2)
    plt.scatter(prox_tbr1_zika[:, 0], prox_tbr1_zika[:, 1], s=1)
    rect = Rectangle((start + stop)/2, stop[0]-start[0], stop[1]-start[1], edgecolor='r', linewidth=1, fill=False)
    ax2.add_patch(rect)
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.title(f'TBR1+ cells in Zika_{zika_idx}')
    plt.xlabel('SOX2 proximity')
    plt.ylabel('TBR1 proximity')
    plt.tight_layout()
    plt.show()
```

```python
def apply_gate(start, stop, prox):
    idx0 = np.where(np.logical_and(prox[:, 0] > start[0], prox[:, 0] < stop[0]))[0]
    idx1 = np.where(np.logical_and(prox[:, 1] > start[1], prox[:, 1] < stop[1]))[0]
    idx = np.intersect1d(idx0, idx1)
    return prox[idx], idx
```

```python
start = np.asarray([0.5, 0.3])
stop = np.asarray([0.6, 0.4])

show_gate(start, stop)

prox_tbr1_mock_gate, idx_gate_mock = apply_gate(start, stop, prox_tbr1_mock)
prox_tbr1_zika_gate, idx_gate_zika = apply_gate(start, stop, prox_tbr1_zika)

prox_tbr1_mock_gate.shape, prox_tbr1_zika_gate.shape
```

```python
syto_mock = io.open(os.path.join(mock_path, 'dataset/syto.zarr'))
sox2_mock = io.open(os.path.join(mock_path, 'dataset/sox2.zarr'))
tbr1_mock = io.open(os.path.join(mock_path, 'dataset/tbr1.zarr'))

syto_zika = io.open(os.path.join(zika_path, 'dataset/syto.zarr'))
sox2_zika = io.open(os.path.join(zika_path, 'dataset/sox2.zarr'))
tbr1_zika = io.open(os.path.join(zika_path, 'dataset/tbr1.zarr'))

syto_mock.shape, syto_zika.shape
```

```python
%matplotlib notebook
```

```python
n = 3
width = 256

sample_idx_mock = np.random.choice(np.arange(len(prox_tbr1_mock_gate)), n**2, replace=False)

plt.figure(figsize=(8, 8))
for i in range(n):
    for j in range(n):
        k = j + i*n
        p_mock = prox_tbr1_mock_gate[sample_idx_mock[k]]
        center = centroids_mock[idx_gate_mock[sample_idx_mock[k]]]
        roi_start = np.asarray([center[0]-1, center[1] - width//2, center[2] - width//2])
        roi_stop = np.asarray([center[0]+1, center[1] + width//2, center[2] + width//2])
        syto_roi = utils.extract_box(syto_mock, roi_start, roi_stop) 
        sox2_roi = utils.extract_box(sox2_mock, roi_start, roi_stop) 
        tbr1_roi = utils.extract_box(tbr1_mock, roi_start, roi_stop) 
        
        roi = np.dstack([sox2_roi.max(axis=0), tbr1_roi.max(axis=0), syto_roi.max(axis=0)])
        
        plt.subplot(n, n, k+1)
        plt.imshow(roi)
        plt.plot([width//2], [width//2], 'y*')
        plt.title(str(center))
        plt.axis('off')
plt.show()
```
