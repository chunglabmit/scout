## Look at cell counts and frequency vs morphology

```python
working_dir
```

```python
dfs = []
for i, org_path in enumerate(analysis.index):
    folder = analysis.loc[org_path]['type']
    if folder == 'Lancaster_d35':
        path = os.path.join(working_dir, folder, org_path, 'individual_ventricle_cellfreq.xlsx')
        df = pd.read_excel(path, index_col=0)
        df['org_name'] = len(df) * [org_path]
        df['org_idx'] = len(df) * [i]
        dfs.append(df)
df = pd.concat(dfs)
df.head()
```

```python
plt.figure(figsize=(8, 18))
plt.subplot(3, 1, 1)
for i in df['org_idx'].unique():
    sns.scatterplot(x='eq_diam', y='sox2_freq', data=df.where(df['org_idx'] == i), label=f'Organoid {i}')
plt.ylim([0, 1])
plt.subplot(3, 1, 2)
for i in df['org_idx'].unique():
    sns.scatterplot(x='eq_diam', y='tbr1_freq', data=df.where(df['org_idx'] == i))
plt.ylim([0, 1])
plt.subplot(3, 1, 3)
for i in df['org_idx'].unique():
    sns.scatterplot(x='eq_diam', y='dn_freq', data=df.where(df['org_idx'] == i))
plt.ylim([0, 1])
plt.show()
```

```python
cols = ['sox2_freq', 'tbr1_freq', 'dn_freq']
colors = ['r', 'g', 'b']
plt.figure(figsize=(8, 6))
for col, color in zip(cols, colors):
    sns.jointplot(x='eq_diam', y=col, data=df, label=col, color=color, kind='reg', scatter_kws={'s': 8, 'alpha': 0.5})
plt.xlim([0, 420])
plt.ylim([0, 1])
plt.ylabel('Frequency')
plt.xlabel('Eq. diameter ($\mu m$)')
plt.legend()
plt.show()
```

```python
data_flatten = {'freq': pd.concat([df['sox2_freq'], df['tbr1_freq'], df['dn_freq']]),
                'type': len(df) * ['SOX2'] + len(df) * ['TBR1'] + len(df) * ['DN'],
                'eq_diam': 3 * list(df['eq_diam'])}
df_flatten = pd.DataFrame(data_flatten)
df_flatten.head()
```

```python
from scipy import stats
```

```python
p = sns.JointGrid(x='eq_diam', y='freq', data=df_flatten)

sns.regplot(x='eq_diam', y='sox2_freq', data=df, ax=p.ax_joint, color='r', scatter_kws={'s': 4, 'alpha': 0.6}, label='SOX2')
sns.regplot(x='eq_diam', y='tbr1_freq', data=df, ax=p.ax_joint, color='g', scatter_kws={'s': 4, 'alpha': 0.6}, label='TBR1')
sns.regplot(x='eq_diam', y='dn_freq', data=df, ax=p.ax_joint, color='b', scatter_kws={'s': 4, 'alpha': 0.6}, label='DN')

sns.distplot(df_flatten['eq_diam'], ax=p.ax_marg_x, color='k')
sns.distplot(df['sox2_freq'], ax=p.ax_marg_y, color='r', vertical=True)
sns.distplot(df['tbr1_freq'], ax=p.ax_marg_y, color='g', vertical=True)
sns.distplot(df['dn_freq'], ax=p.ax_marg_y, color='b', vertical=True)

p.ax_joint.set_xlim([0, 300])
p.ax_joint.set_ylim([0, 1])
p.ax_joint.legend()
p.ax_marg_x.set_title("{} ventricles, {} organoids".format(len(df), len(df['org_idx'].unique())))
p.ax_joint.set_xlabel('Eq. diameter ($\mu m$)')
p.ax_marg_x.set_xlabel(None)
p.ax_joint.set_ylabel('Frequency')
p.ax_marg_y.set_ylabel(None)

# Get pearson correlations

r_sox2, p_sox2 = stats.pearsonr(df['eq_diam'], df['sox2_freq'])
r_tbr1, p_tbr1 = stats.pearsonr(df['eq_diam'], df['tbr1_freq'])
r_dn, p_dn = stats.pearsonr(df['eq_diam'], df['dn_freq'])

p.ax_joint.annotate('$r_{SOX2}=$' + f'{r_sox2:0.2f}, $p=$ {p_sox2:.1e}', [80, 0.96])
p.ax_joint.annotate('$r_{TBR1}=$' + f'{r_tbr1:0.2f}, $p=$ {p_tbr1:.1e}', [80, 0.92])
p.ax_joint.annotate('$r_{DN}=$' + f'{r_dn:0.2f}, $p=$ {p_dn:.1e}', [80, 0.88])

plt.show()
```

    - SOX2, TBR1, DN counts correlated with organoid and ventricle size
    - SOX2 is especially central in the graph, which makes sense developmentally
