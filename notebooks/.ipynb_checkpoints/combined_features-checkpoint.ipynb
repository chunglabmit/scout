```python
import os
import numpy as np
import pandas as pd
```

```python
working_dir = '/media/jswaney/Drive/Justin/organoid_phenotyping'
os.listdir(working_dir)
```

```python
df = pd.read_excel(os.path.join(working_dir, 'combined_features.xlsx'))#.transpose()
df.head()
```

```python
all_features = []
all_zscores = []
for index, row in df.iterrows():
    feature = row.iloc[0]
    values = row.iloc[1:]
    mean = values.mean()
    stdev = values.std()
    if mean == np.nan or stdev == np.nan:
        continue
    zscores = (values - mean) / stdev
    all_features.append(feature)
    all_zscores.append(zscores)
    print(zscores)
    
```

```python
pd.DataFrame(all_zscores, index=all_features)
```
