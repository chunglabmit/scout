```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
working_dir = '/media/jswaney/Drive/Justin/organoid_phenotyping/datasets/d35_wt'
datasets = os.listdir(working_dir)
datasets
```

```python
proximities = np.load(os.path.join(working_dir, 'niche_proximities_combined.npy'))
x_tsne = np.load(os.path.join(working_dir, 'niche_tsne_combined.npy'))
proximities.shape, x_tsne.shape
```

```python
organoid = np.concatenate([np.ones(5000), 2*np.ones(5000), 3*np.ones(5000)]).astype(np.int)
organoid
```

```python
fig = plt.figure(figsize=(12, 8))
for i in range(3):
    idx = np.where(organoid == i + 1)[0]
    plt.plot(x_tsne[idx, 0], x_tsne[idx, 1], '.', alpha=0.3, markersize=5, label=f'Organoid {i+1}')
plt.legend()
plt.show()
fig.savefig('tsne_3orgs.png')
```

```python
labels = np.load(os.path.join(working_dir, 'niche_labels_combined.npy'))
labels.shape
```

```python
clusters = ['Non-proximal', 'SOX2 adjacent', 'TBR1 adjacent', 'Intermediates']
for i in range(3):
    idx = np.where(organoid == i + 1)[0]
    lbls = labels[idx]
    n = len(lbls)
    print(f'Organoid {i}')
    for j in range(4):
        count = len(np.where(lbls == j)[0])
        print(f'{count / n * 100:.2f}% {clusters[j]}')
    print('')
```
