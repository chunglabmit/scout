```python
import os
from scout.preprocess import histogram
from scout import io
from tqdm import tqdm_notebook as tqdm
import matplotlib.pyplot as plt
```

```python
working_dir = '/data/datasets/organoid_phenotyping/datasets/'
paths = [os.path.join(working_dir, p) for p in os.listdir(working_dir)]
datasets = [p for p in paths if os.path.isdir(p)]
datasets.sort()
len(datasets)
```

# Check middle Syto slice histogram

```python
hists = []
centers = []
for dataset in tqdm(datasets, total=len(datasets)):
    arr = io.open(os.path.join(dataset, 'syto.zarr'))
    img = arr[len(arr) // 2]
    hist, center = histogram(img)
    hists.append(hist)
    centers.append(center)
    plt.plot(center, hist / hist.sum())
    plt.title(os.path.basename(dataset))
    plt.xlim([0, 1])
    plt.ylim([0, 0.04])
    plt.show()
```

Last d35 and first d60 have issues with Zarr version of syto channel
