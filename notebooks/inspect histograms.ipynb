```python
folders = [os.path.join(working_dir, p) for p in datasets]
```

```python
os.listdir(folders[0])
```

```python
import pandas as pd
```

```python
hists = [os.path.join(f, 'Ex0_hist.csv') for f in folders]
dfs = [pd.read_csv(h) for h in hists]
dfs[0].head()
```

```python
import matplotlib.pyplot as plt
```

```python
for df in dfs:
    x = df['intensity']
    y = df['count']
    mean = (x * y).sum() / y.sum()
    print(mean)
    plt.plot(x, y)
    plt.plot([mean, mean], [0, 1e6], 'k-')
plt.xlim([0, 1000])
plt.ylim([0, 1e6])
plt.show()
```
